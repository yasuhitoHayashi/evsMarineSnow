```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from datetime import datetime
import seaborn as sns

sns.set(
    "paper",
    "whitegrid",
    "bright",
    font_scale=1.5,
    rc={"lines.linewidth": 1, "grid.linestyle": "--"},
)

# ---------------------------
# 各カメラのCSVファイルが保存されているディレクトリを定義
# ---------------------------
directories = {
    "cam1": Path('/Users/yasu-air/Dropbox/yhLab/KS24-16/cam1/'),
    "cam2": Path('/Users/yasu-air/Dropbox/yhLab/KS24-16/cam2/')
}

# 各カメラのプロット色を設定（例：cam1は黒、cam2は赤）
colors = {
    "cam1": "black",
    "cam2": "red"
}

# プロット用のFigureを作成
plt.figure(figsize=(10, 6))

# 各ディレクトリごとに処理
for cam_name, directory_path in directories.items():
    # CSVファイルを取得（例：末尾が _noWarp.csv のもの）
    csv_files = list(directory_path.glob('*_noWarp.csv'))
    
    # 回帰用に全ファイルのデータ点をまとめるリストを初期化
    x_all = []
    y_all = []
    
    for csv_file in csv_files:
        try:
            print(f"{csv_file.name} （{cam_name}） を処理中...")
            
            # CSVファイルを読み込み
            df = pd.read_csv(csv_file)
            
            # x軸用の列を決定（"Major_Axis_noWarp" があればそれを、なければ "Major_Axis" を使用）
            if 'Major_Axis_noWarp' in df.columns:
                x_vals = df['Major_Axis_noWarp'] * 0.156
            elif 'Major_Axis' in df.columns:
                x_vals = df['Major_Axis'] * 0.156
            else:
                print(f"{csv_file.name} に x 軸用のカラムが見つかりません。スキップします。")
                continue
            
            # y軸は eventCount 列
            if 'eventCount' not in df.columns:
                print(f"{csv_file.name} に 'eventCount' 列が存在しません。スキップします。")
                continue
            y_vals = df['eventCount']
            
            # サイズが1mm以上のデータに限定する
            mask = x_vals >= 1
            x_vals_filtered = x_vals[mask]
            y_vals_filtered = y_vals[mask]
            
            if len(x_vals_filtered) == 0:
                print(f"{csv_file.name} には1mm以上のデータがありません。スキップします。")
                continue
            
            # 散布図をプロット（αをつけて重なりを軽減）
            plt.scatter(
                x_vals_filtered,
                y_vals_filtered,
                label=f'{cam_name}_{csv_file.stem}',
                color=colors[cam_name],
                alpha=0.5
            )
            
            # 回帰用にデータを追加
            x_all.extend(x_vals_filtered.tolist())
            y_all.extend(y_vals_filtered.tolist())
            
        except Exception as e:
            print(f"{csv_file.name} の処理中にエラーが発生しました: {e}\n")
    
    # 各ディレクトリ内でデータが存在する場合、線形フィッティングを実施
    if len(x_all) > 0:
        x_all_arr = np.array(x_all)
        y_all_arr = np.array(y_all)
        # 1次の線形フィッティング（最小二乗法）
        slope, intercept = np.polyfit(x_all_arr, y_all_arr, 1)
        # x の範囲をカバーする直線を作成
        x_fit = np.linspace(x_all_arr.min(), x_all_arr.max(), 100)
        y_fit = slope * x_fit + intercept
        # フィッティングラインをプロット
        plt.plot(
            x_fit,
            y_fit,
            color=colors[cam_name],
            linewidth=2,
            label=f'{cam_name} fit'
        )
    else:
        print(f"{cam_name} のデータが見つかりませんでした。")

# 軸ラベル、グリッド、凡例の設定
plt.xlabel('Major axis length (mm)')
plt.ylabel('eventCount')
plt.grid(True, which="both", ls="--", linewidth=0.5)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
from datetime import datetime
import seaborn as sns
from scipy.stats import ttest_ind

# Seaborn のスタイル設定
sns.set(
    "paper",
    "whitegrid",
    "bright",
    font_scale=1.5,
    rc={"lines.linewidth": 1, "grid.linestyle": "--"},
)

# 各カメラのCSVファイルが保存されているディレクトリのパスを定義
directories = {
    "cam1": Path('/Users/yasu-air/Dropbox/yhLab/KS24-16/cam1/'),
    "cam2": Path('/Users/yasu-air/Dropbox/yhLab/KS24-16/cam2/')
}

# 結果をまとめるリスト
data_list = []

# 各カメラごとに処理
for cam_name, directory_path in directories.items():
    csv_files = list(directory_path.glob('*_noWarp.csv'))
    
    for csv_file in csv_files:
        try:
            print(f"{csv_file.name} （{cam_name}） を処理中...")
            df = pd.read_csv(csv_file)
            
            # 必要な列が存在するかチェック
            if 'Major_Axis' not in df.columns or 'eventCount' not in df.columns:
                print(f"{csv_file.name} に必要な列がありません。スキップします。\n")
                continue
            
            # CSVファイルが1粒子の測定結果であると仮定し、最初の行の値を採用
            major_axis = df['Major_Axis'].iloc[0] * 0.156  # mm 単位に変換
            event_count = df['eventCount'].iloc[0]
            
            # 正規化イベント数：粒子サイズに対するイベント数
            norm_event = event_count / major_axis
            
            data_list.append({
                "camera": cam_name,
                "normalized_event": norm_event,
                "major_axis_mm": major_axis
            })
        except Exception as e:
            print(f"{csv_file.name} の処理中にエラーが発生しました: {e}")

# DataFrameにまとめる
df_results = pd.DataFrame(data_list)
print("\n集計結果:")
print(df_results)

# ---------------------------
# 有意差検定（Welch の t 検定）
# ---------------------------
# 各カメラごとの正規化イベント数のデータを抽出
cam1_data = df_results[df_results['camera'] == 'cam1']['normalized_event']
cam2_data = df_results[df_results['camera'] == 'cam2']['normalized_event']

t_stat, p_val = ttest_ind(cam1_data, cam2_data, equal_var=False)
print("\n有意差検定 (Welch の t 検定) 結果:")
print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_val:.3f}")

# ---------------------------
# 箱ひげ図で各カメラごとの正規化イベント数の分布をプロット
# ---------------------------
plt.figure(figsize=(8, 6))
sns.boxplot(x="camera", y="normalized_event", data=df_results)
plt.xlabel("Camera")
plt.ylabel("Normalized Event Count (events per mm)")
plt.grid(True, which="both", ls="--", linewidth=0.5)

# 有意差検定の結果を図に注釈として追加
plt.text(0.5, max(df_results['normalized_event'])*0.95, f'p = {p_val:.3f}',
         horizontalalignment='center', fontsize=14)

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
import seaborn as sns
from scipy.stats import ttest_ind

# Seaborn のスタイル設定
sns.set(
    "paper",
    "whitegrid",
    "bright",
    font_scale=1.5,
    rc={"lines.linewidth": 1, "grid.linestyle": "--"},
)

# 各カメラのCSVファイルが保存されているディレクトリのパスを定義
directories = {
    "40m": Path('/Users/yasu-air/Dropbox/yhLab/KS24-16/cam1/'),
    "80m": Path('/Users/yasu-air/Dropbox/yhLab/KS24-16/cam2/')
}

# 結果をまとめるリスト
data_list = []

# 各カメラごとにCSVファイルを処理
for cam_name, directory_path in directories.items():
    csv_files = list(directory_path.glob('*_noWarp.csv'))
    
    for csv_file in csv_files:
        try:
            print(f"{csv_file.name} （{cam_name}） を処理中...")
            df = pd.read_csv(csv_file)
            
            # 必要な列が存在するかチェック
            if 'Projected_Points_Size' not in df.columns or 'eventCount' not in df.columns:
                print(f"{csv_file.name} に必要な列がありません。スキップします。\n")
                continue
            
            # CSVファイルが1粒子の測定結果であると仮定し、最初の行の値を採用
            # 単位変換は行わず、そのまま使用
            projected_size = df['Projected_Points_Size'].iloc[0]
            event_count = df['eventCount'].iloc[0]
            
            # 正規化イベント数：粒子サイズに対するイベント数
            norm_event = event_count / projected_size
            
            data_list.append({
                "camera": cam_name,
                "normalized_event": norm_event,
                "projected_size": projected_size
            })
        except Exception as e:
            print(f"{csv_file.name} の処理中にエラーが発生しました: {e}")

# DataFrame にまとめる
df_results = pd.DataFrame(data_list)
print("\n全体の集計結果:")
print(df_results)

# ---------------------------
# 有意差検定 (Welch の t 検定)
# ---------------------------
cam1_data = df_results[df_results['camera'] == '40m']['normalized_event']
cam2_data = df_results[df_results['camera'] == '80m']['normalized_event']

t_stat, p_val = ttest_ind(cam1_data, cam2_data, equal_var=False)
print("\n有意差検定 (Welch の t 検定) 結果:")
print(f"T-statistic: {t_stat:.3f}")
print(f"P-value: {p_val:.3f}")

# ---------------------------
# 箱ひげ図で各カメラごとの正規化イベント数の分布をプロット
# ---------------------------
plt.figure(figsize=(8, 6))
sns.boxplot(x="camera", y="normalized_event", data=df_results)
plt.ylabel("Normalized Event Count (events per unit size)")
plt.grid(True, which="both", ls="--", linewidth=0.5)

# 最大値を取得して注釈の y 座標を決定
max_y = df_results["normalized_event"].max()
plt.text(0.5, max_y * 0.95, f'p = {p_val:.3f}', horizontalalignment='center', fontsize=14)

plt.show()
```
